```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression,SGDRegressor
from sklearn.preprocessing import PolynomialFeatures,StandardScaler
from sklearn.metrics import r2_score
```

```python


# Generate 30 values for X1
X = np.linspace(-10, 10, 100).reshape(-1, 1)

# Generate target y with quadratic relationship and noise
y = 0.8 * X**2 + 0.9 * X + 2 + np.random.randn(100, 1)
```

```python
plt.figure(figsize=(15,6))
plt.plot(X, y,'b.')
plt.xlabel("X")
plt.ylabel("y")
plt.show()
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=2)
```

```python
# Applying linear regression
lr = LinearRegression()
```

```python
lr.fit(X_train,y_train)
```

```python
y_pred = lr.predict(X_test)
r2_score(y_test,y_pred)
```

```python
plt.plot(X_train,lr.predict(X_train),color='r')
plt.plot(X, y, "b.")
plt.xlabel("X")
plt.ylabel("y")
plt.show()
```

```python
poly=PolynomialFeatures(degree=2,include_bias=True)
X_train_trns=poly.fit_transform(X_train)
X_test_trsn=poly.transform(X_test)
```

```python
print(X_train[0])
print(X_train_trns[0])
```

```python
lr=LinearRegression()
lr.fit(X_train_trns,y_train)
```

```python
y_pred=lr.predict(X_test_trsn)
```

```python
r2_score(y_test,y_pred)
```

```python
print(lr.coef_)
print(lr.intercept_)
```

```python
plt.figure(figsize=(15,6))
plt.plot(X ,y, "y-", linewidth=2, label="Predictions")
plt.plot(X_train, y_train, "g.",label='Training points')
plt.plot(X_test, y_test, "r.",label='Testing points')
plt.xlabel("X")
plt.ylabel("y")
plt.legend()
plt.show()
```
